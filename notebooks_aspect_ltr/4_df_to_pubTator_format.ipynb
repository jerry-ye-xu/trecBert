```python
import csv

import pandas as pd
```

```python
df = pd.read_pickle("../data/all_ct_info_labelled.pickle")
df.head()
```

```python
df.id.nunique()
```

```python
df.shape
```

```python
pub = df.melt(
    id_vars="id", 
    value_vars=["brief_title", "brief_summary"],
)
pub["variable"] = pub["variable"].str.replace("brief_title", "t")
pub["variable"] = pub["variable"].str.replace("brief_summary", "a")
pub[pub["id"] == "NCT02567942"]
```

```python
pub.sort_values(by=["id", "variable"], ascending=False, inplace=True)
pub.head(n=10)
```

```python
df_path = "../debug_data/pubTator_format.csv"

pub.to_csv(df_path, sep="|", header=False, index=False)
```

```python
pub_reload = pd.read_csv(
    df_path, sep="|", 
    header=None, names=["doc_id", "section", "text"])
pub_reload.head(n=10)
```

```python
df_space_path = "../debug_data/pubTator_format_space.csv"

with open(df_path, 'r') as infile:
    pub = csv.reader(infile, delimiter='|')
    with open(df_space_path, 'wt') as outfile:
        outwriter = csv.writer(outfile, delimiter='|')
        i = 0
        for row in pub:
            outwriter.writerow(row)
            i += 1
            if i % 2 == 0:
                # write the empty row
#                 outwriter.writerow([])
                outwriter.writerow([])
```

```python
pub_reload = pd.read_csv(
    df_space_path, sep="|", 
    header=None, names=["doc_id", "section", "text"])
pub_reload.head(n=10)
```
